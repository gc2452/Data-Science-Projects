## PIVOT TABLE THAT ORGANIZES DIFF SERIES BY COUNTRY

```python
table = pd.pivot_table(df, values='2014 [YR2014]', index=['Country Name'], columns=['Series Name'], aggfunc=np.sum)
```

```python
table
```

## RESET THE INDEX

```python
tablereset = table.reset_index(level=None, drop=False, inplace=False, col_level=0, col_fill='')
```

```python
tablereset['GDP (current US$)'] = tablereset['GDP (current US$)'].apply(pd.to_numeric,errors='coerce')
```

```python
tablereset2 = tablereset['GDP (current US$)'].dropna()
```

## CONVERT TYPE OBJECT TO NUMERIC

```python
tablereset2['GDP (current US$)'] = tablereset2['GDP (current US$)'].apply(pd.to_numeric,errors='coerce') #uses pandas
```

```python
tablereset2['GDP (current US$)'] = tablereset2['GDP (current US$)'].convert_objects(convert_numeric=True) #uses python
```
